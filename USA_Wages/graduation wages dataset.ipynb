```python
Tec_univ['Data'] = Tec_univ['Data'].astype('datetime64[ns]')
```

```python
# Inserir valores referentes ao ano de análise
```

```python
Tec_univ["Ano"] = Tec_univ['Data'].dt.year
```

```python
# Para apagar uma coluna
# Tec_univ = Tec_univ.drop(columns=['Data'])
```

```python
# Tirar a média por Ano
```

```python
media = Tec_univ.groupby("Ano").mean()
```

```python
media
```

```python
# Criação de Gráficos
```

```python
media.plot()
plt.title('Tecnicos vs Universitários')
plt.ylabel('Porcentagem')
plt.xlabel('Ano')
plt.show()
```
